## Визуализация магнитного поля произвольных контуров с постоянным током.

**Цель работы:** сделать наглядную визуализацию магнитного поля от контура с постоянным током.

**В работе используются:** закон Био--Савара--Лапласа.

**Мотивация:** Важно уметь рассчитывать магнитное поле в различных точках пространства, например, при построении микросхем. Но уже в таком простом случае, как кольцо с током, при нахождении точки наблюдения не на оси кольца не существует аналитического результата, поэтому высока роль численного моделирования.

**Теоретическое обоснование.**
Мы прекрасно умеем считать поле от прямолинейного проводника с током. Так давайте же выведем эту формулу!

В данной точке элемент с током $I$ длины $dl$ и радиус-вектором начала $\overrightarrow{r}$ по закону Био-Савара-Лапласа создаёт магнитное поле:
$$\overrightarrow{dB} = \frac{\mu_0}{4\pi} \mu I \cdot \frac{[\overrightarrow{dl} \times \overrightarrow{r}]}{r^3}.$$

Из рисунка легко увидеть, что:
$$r = \frac{R}{\cos{\phi}}, dl = \frac{r \cdot d\phi}{\cos{\phi}} \Longrightarrow dB = \frac{\mu_0}{4\pi} \mu I \cdot \frac{\cos{\phi}}{R}d\phi.$$

Назовём буквой $\alpha$ угол $90^{\circ} + \phi$. Пусть для определённости концы проводника видны из данной точки под углами $\alpha_1 \text{ и } \alpha_2$ и первый конец ближе к точке наблюдения, тогда имеем:
$$B = \int_{\alpha_1}^{\alpha_2} \frac{\mu_0}{4\pi} \mu I \cdot \frac{\sin{\alpha}}{R}d\alpha = \frac{\mu_0 \mu I}{4\pi R} \cdot (\cos{\alpha_1} - \cos{\alpha_2}).$$

Эта формула является основной. С помощью неё рассчитываются все контуры, заданные в визуализации (кольцевой контур приближается многоугольником). 

**Замечание.** Покажем, что для кольцевого контура при смещении точки с его оси магнитное поле не считается аналитически.

Рассмотрим точку с координатами $(z, y, 0)$. Тогда для $\overrightarrow{dl}, \overrightarrow{r}$ и их векторного произведения имеем:
$$\overrightarrow{dl} = \left(\begin{array}{@{}c@{}}
    -R\cos{\phi} \\
    -R\sin{\phi} \\
    0
    \end{array} \right) \cdot d\phi, \overrightarrow{r} = \left(\begin{array}{@{}c@{}}
    R\sin{\phi} \\
    -(R\cos{\phi} - y) \\
    z
    \end{array} \right), [\overrightarrow{dl} \times \overrightarrow{r}] = \left(\begin{array}{@{}c@{}}
    -Rz\sin{\phi} \\
    Rz\cos{\phi} \\
    R(R - y\cos{\phi})
    \end{array} \right) \cdot d\phi$$

Тогда для компонент вектора магнитной индукции имеем:
$$B_x(y, z) = -\frac{\mu_0 \mu IRz}{4\pi}\int_0^{2\pi} \frac{\sin{\phi}}{(R^2 + y^2 - 2Ry\cos{\phi} + z^2)^{3/2}} d\phi$$
$$B_y(x, z) = \frac{\mu_0 \mu IRz}{4\pi}\int_0^{2\pi} \frac{\cos{\phi}}{(R^2 + y^2 - 2Ry\cos{\phi} + z^2)^{3/2}} d\phi$$
$$B_z(x, y) = \frac{\mu_0 \mu IR}{4\pi}\int_0^{2\pi} \frac{R - y\cos{\phi}}{(R^2 + y^2 - 2Ry\cos{\phi} + z^2)^{3/2}} d\phi$$

Первый интеграл обращается в 0 в следствие симметрии, остальные два интеграла не имеют аналитического результата и могут быть найдены только численно.

**Визуализация.** Стоит отметить, пару моментов.

1) Для визуализации специльно были выбраны **безынертные** частицы, а не векторные поля или линии магнитного поля. Оба последних варианта хорошо показывают направления, но о величине магнитного поля визуально говорят весьма мало (как правило, с помощью цвета или вообще никак), а вот скорость частиц хорошо отражает величину магнитного поля. (Безынертность нужна для того, чтобы частицы не искажали картину в следствие своей массы.)

2) Визуализация даёт **относительные величины** магнитного поля, так как расстояние мерилось в пикселях, а для получения скорости частиц из $\overrightarrow{B}$ использовался специально подобранный коэффициент.
